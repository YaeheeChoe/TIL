그래프 도구 커스터마이징

```python
def graph_init():
    plt.figure(figsize=(8,8))
    plt.xlabel('CCTV 설치개수')
    plt.ylabel('동네')
cctv = ps.read_csv('cctv.csv')
```

선형그래프

```python

def display_line_graph():
    graph_init()
    cctvGraph = cctv.drop(index=0,axis=0)
    plt.plot(cctvGraph.apply(lambda x:int(x['총계'].replace(',','')),axis=1),cctvGraph['구분'],marker='*',color='blue')
    plt.show()
    return
display_line_graph()
```

막대그래프

```python
def display_barh_graph():
     graph_init()
     gdata = cctv.sort_values(by="총계",ascending=True)
     gdata = gdata.drop(index=0,axis=0)
     plt.barh(gdata['구분'],gdata['총계'].apply(lambda x:int(x.replace(',',''))),axis=1),color='gold')
     # 람다 일단 뭔말인진 모르겠고
     plt.show()
display_barh_graph()
```
